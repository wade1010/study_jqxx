#### 需求：画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度~18度

```python
import random
```

```python
# 0 准备数据
x = range(60)
y_bj = [random.uniform(15,18) for i in x]

# 1 创建画布
plt.figure(figsize=(20,8),dpi=100)

# 2 绘制图像
plt.plot(x,y_bj)

# 3 显示
plt.show()
```

## 实现一些其他功能

```python
plt.rcParams['font.sans-serif'] = 'Noto Sans CJK JP'
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
```

```python
# 0 准备数据
x = range(60)
y_bj = [random.uniform(15,18) for i in x]
y_sh = [random.uniform(2,5) for i in x]

# 1 创建画布
plt.figure(figsize=(20,8),dpi=100)

# 2 绘制图像
plt.plot(x,y_bj,label='北京')
plt.plot(x,y_sh,color='r',linestyle='--',label='上海')

# 2.1 添加x,y轴
x_ticks_label = ["11点{}分".format(i) for i in x]


# 2.2修改x,y轴坐标刻度显示
plt.xticks(x[::5],x_ticks_label[::5])
y_ticks = range(40)
plt.yticks(y_ticks[::5])

# 2.3 添加网格
plt.grid(True,linestyle='--',alpha=0.8)

# 2.4 添加x,y轴描述信息及标题
plt.xlabel = '时间'
plt.ylabel = '温度'
plt.title('中午11点0分到12点之间的温度变化图示',fontsize=22)

plt.legend(loc=0)

# 2.5 图像保存  注意：plt.show()会释放figure资源，如果在显示图像之后保存图片将只能保存空图片。
plt.savefig('./plt_saved.png')

# 2.6 增加上海的温度

# 3 显示
plt.show()
```
