```python
data = pd.read_csv("./dataset/stock_day.csv")
```

```python
data.head()
```

```python
p_change = data["p_change"]
```

```python
p_change.head()
```

```python
# 自动分成差不多数量的类别
qcut = pd.qcut(p_change, 10)
qcut.value_counts()
```

```python
# 指定分组区间
bins = [-100,-7-5,-3,0,3,5,7,100]
p_count = pd.cut(p_change,bins)
```

```python
p_count.value_counts()
```

```python
dummies = pd.get_dummies(p_count,prefix='rise')
dummies.head()
```

# 交叉透视表

```python
data.head()
```

```python
data.index
```

```python
time = pd.to_datetime(data.index)
```

```python
time.weekday
```

```python
time.day
```

```python
data['week'] = time.weekday
```

```python
data.head()
```

```python
data['p_n'] = np.where(data['p_change']>0,1,0)
```

```python
data.head()
```

```python
count = pd.crosstab(data['week'],data['p_n']) #计算一列数据对于另外一列数据的分组个数
count
```

```python
# 算数运算，先求和
sum = count.sum(axis=1).astype(np.float32)
```

```python
sum
```

```python
# 进行相除操作，得出比例
pro = count.div(sum, axis=0)
```

```python
pro
```

```python
#使用plot画出这个比例，使用stacked的柱状图
pro.plot(kind='bar', stacked=True)
plt.show()
```

```python
#使用pivot_table实现
pro2=data.pivot_table(['p_n'],index='week')
pro2
```

## 分组聚合

```python
col =pd.DataFrame({'color': ['white','red','green','red','green'], 
                   'object': ['pen','pencil','pencil','ashtray','pen'],
                   'price1':[5.56,4.20,1.30,0.56,2.75],'price2':[4.75,4.12,1.60,0.75,3.15]})

col
```

```python
col.groupby(['color'])['price1'].mean()
```

```python
col['price1'].groupby(col['color']).mean()
```

```python
col.groupby(['color'],as_index=False)['price1'].mean()
```

## 星巴克零售店铺数据

```python
starbucks = pd.read_csv('./dataset/starbucks.csv')
```

```python
starbucks.head()
```

```python
# 按照国家分组，求出每个国家的星巴克零售店数量
count = starbucks.groupby('Country').count()
count
```

```python
count['Brand'].plot(kind='bar',figsize=(20,8))
plt.show()
```

```python
# 假设我们加入省市一起进行分组 设置多个索引，set_index()
count2 = starbucks.groupby(['Country', 'State/Province']).count()
count2
```

```python
count2.head(10)['Brand'].plot(kind='bar',figsize=(20,8))
plt.show()
```
