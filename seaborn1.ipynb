```python
import seaborn as sns
import numpy as np
```

# 绘制单变量分布

```python
np.random.seed(0)  #确定随机成成的数据是一样的
arr = np.random.randn(100)
# sns.displot(arr,kind='hist',rug=True) 
sns.distplot(arr,bins=10,hist=True,kde=True,rug=True)  #未来会废弃 
```

# 绘制双变量分布

## 绘制散点图

```python
import pandas as pd
```

```python
df = pd.DataFrame({'x':np.random.randn(500),'y':np.random.randn(500)})
df.head()
```

```python
sns.jointplot(data=df,kind='scatter')
```

## 绘制二维直方图

```python
sns.jointplot(x="x", y="y", data=df, kind="hex")
```

## 绘制核密度估计图形

```python
sns.jointplot(x="x", y="y", data=df, kind="kde")
```

## 绘制成对的双变量分布

```python
# 加载seaborn中的数据集
dataset = sns.load_dataset("iris")

dataset.head()
```

```python
# 绘制多个成对的双变量分布
sns.pairplot(dataset)
```
