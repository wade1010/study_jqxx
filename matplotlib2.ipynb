```python
import matplotlib.pyplot as plt
import random
```

```python
plt.rcParams['font.sans-serif'] = 'Noto Sans CJK JP'
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
```

```python
x = range(60)
y_sh = [random.uniform(1,5) for i in x]
y_bj = [random.uniform(15,18) for i in x]

fig , axes = plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=100)

axes[0].plot(x,y_bj,label='北京')
axes[1].plot(x,y_sh,label='上海',color='r',linestyle='--')

x_ticks_label = ["11点{}分".format(i) for i in x]
y_ticks = range(40)

axes[0].set_xticks(x[::5])
axes[0].set_yticks(y_ticks[::5])
axes[0].set_xticklabels(x_ticks_label[::5])
axes[0].set_title('中午11点0分到12点之间的温度变化图示',fontsize=22)

axes[1].set_xticks(x[::5])
axes[1].set_yticks(y_ticks[::5])
axes[1].set_xticklabels(x_ticks_label[::5])
axes[1].set_title('中午11点0分到12点之间的温度变化图示',fontsize=22)


axes[0].grid(True,linestyle='--',alpha=0.9)
axes[1].grid(True,linestyle='--',alpha=0.9)

plt.savefig('./plt_saved2.png')

axes[0].legend(loc=0)
axes[1].legend(loc=0)
plt.show()
```
