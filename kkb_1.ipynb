```python
import numpy as np
```

```python
def func(x):
    return 10*x**2 + 37*x + 9
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10)
plt.plot(x,func(x))
```

```python
import random
```

```python
def gradien(x):
    return 20*x+37
```

```python
steps = []
x_optimal = random.choice(x)
x_optimal
```

```python
alpha = 1e-4
for i in range(300):
    x_optimal = x_optimal + (-1)*gradien(x_optimal)*alpha
    steps.append(x_optimal)
for s in steps:
    print(s,func(s))
```
